##### mRNA
DNA-molekylet befinner seg i cellekjernen, men selve byggeprosessen skjer utenfor cellekjernen, p√• *ribosomene*. Man flytter ikke DNA-molekylet ut, men lager en kopi av den ene siden i DNA. Kopien kaller vi for **mRNA**. Det er et enzym som heter **RNA-polymerase** leser av DNA og lager mRNA. Den g√•r langs en spesifikk side av DNA-molekylet.

```
DNA:
TACCTATCTGGACCAACT
```

N√•r **RNA-polymerase** skal kopiere dette segmentet s√• tar vi de motsatte baseparene, men basen Tymin byttes ut med Uracil.

- A -> U
- T -> A
- C -> G
- G -> C

Kopien av DNA-molekylet over ser da slik ut.

```

AUGGAUAGACCUGGUUGA
```

Dette mRNA-molekylet sendes s√• ut av cellekjernen og til ribosomene. Der leses det av tre og tre baser, noe som heter *tripletter* eller *kodoner*.

```

Tilsvarer aminosyrene:

```

En triplett (AUG) koder for starten av proteinet. Denne koder ogs√• for metionin. Det betyr alts√• at alle proteiner starter med √• ha metionin som sin f√∏rste aminosyre, men noen ganger blir deler kuttet av f√∏r proteinet er helt ferdig, s√• ikke alle proteiner i kroppen starter med metionin. 

Det er ogs√• tre av triplettene som koder for et stoppsignal (UAA, UAG, UGA). Da f√•r man ikke en ny aminosyre, men det signaliseres at proteinet er ferdig.

Proteiner best√•r vanligvis av hundrevis og noen ganger tusenvis av aminosyrer. La oss gjennomf√∏re hele proteinsyntesen med Python üêç

````{admonition} Hvilken side skal man lese av?
:type: note
Hver side av DNA-tr√•den har en start og en slutt. Den ene siden kalles for `5'` og den andre siden kalles for `3'`. Navnet kommer av hvor fosfatgruppen er bundet til sukkermolekylet. `5'`-siden ender med en fosfatgruppe, og `3'`-siden ender med en sukkergruppe.

```

```

RNA-polymerase leser fra `3'` til `5'` slik som alle eksemplene p√• denne siden, men legg merke til at mange l√¶reb√∏ker og ressurser p√• internett ofte skriver DNA-molekyler i retningen `5' -> 3'`. Hvis man har en DNA-tr√•d skrevet med `5' -> 3'` holder det faktisk bare √• bytte ut `T` med `U` for √• f√• `mRNA`.
````

#### Programmering av proteinsyntesen

##### Del 1: Lage mRNA

Vi kan representere en bit av DNA som en string i Python. M√•let v√•rt er √• lage en tr√•d med mRNA fra basene i DNA. For √• oversette basene bruker vi en ordbok.

``````python
# En DNA-tr√•d, representert som en tekst.
DNA = "TACCTATCTGGACCAACT"

# Ordbok som oversetter fra baser i DNA til baser i mRNA
DNA_til_mRNA = {
    "A" : "U",
    "T" : "A",
    "C" : "G",
    "G" : "C"
}

# En tom mRNA-Tr√•d
mRNA = ""

# Tar alle basene i DNA, oversetter, og setter inn i mRNA-tr√•den
for x in DNA:
    mRNA += DNA_til_mRNA[x]

# Skriver ut DNA og mRNA
print("DNA:  ", DNA)
print("mRNA: ", mRNA)
``````

##### Del 2: Lage aminosyrer

N√• gjenst√•r det bare √• lage aminosyrer fra triplettene p√• mRNA-tr√•den. For √• oversette bruker vi en *kodontabell*, en oversikt over hvilke tripletter som tilsvarer hvilke aminosyrer.

``````python
# En DNA-tr√•d, representert som en tekst.
DNA = "TACCTATCTGGACCAACT"

# Ordbok som oversetter fra baser i DNA til baser i mRNA
DNA_til_mRNA = {
    "A" : "U",
    "T" : "A",
    "C" : "G",
    "G" : "C"
}

# En tom mRNA-Tr√•d
mRNA = ""

# Tar alle basene i DNA, oversetter, og setter inn i mRNA-tr√•den
for x in DNA:
    mRNA += DNA_til_mRNA[x]

# Skriver ut DNA og mRNA
print("DNA:  ", DNA)
print("mRNA: ", mRNA)

# Ordbok som oversetter fra tripletter til aminosyrer.
kodontabell = {
    "UUU" : "Fenylalanin",     "UCU" : "Serin",    "UAU" : "Tyrosin",       "UGU" : "Cystein",
    "UUC" : "Fenylalanin",     "UCC" : "Serin",    "UAC" : "Tyrosin",       "UGC" : "Cystein",
    "UUA" : "Leucin",          "UCA" : "Serin",    "UAA" : "STOPP",         "UGA" : "STOPP",
    "UUG" : "Leucin",          "UCG" : "Serin",    "UAG" : "STOPP",         "UGG" : "Tryptofan",
    "CUU" : "Leucin",          "CCU" : "Prolin",   "CAU" : "Histidin",      "CGU" : "Arginin",
    "CUC" : "Leucin",          "CCC" : "Prolin",   "CAC" : "Histidin",      "CGC" : "Arginin",
    "CUA" : "Leucin",          "CCA" : "Prolin",   "CAA" : "Glutamin",      "CGA" : "Arginin",
    "CUG" : "Leucin",          "CCG" : "Prolin",   "CAG" : "Glutamin",      "CGG" : "Arginin",
    "AUU" : "Isoleucin",       "ACU" : "Threonin", "AAU" : "Asparagin",     "AGU" : "Serin",
    "AUC" : "Isoleucin",       "ACC" : "Threonin", "AAC" : "Asparagin",     "AGC" : "Serin",
    "AUA" : "Isoleucin",       "ACA" : "Threonin", "AAA" : "Lysin",         "AGA" : "Arginin",
    "AUG" : "Metionin(START)", "ACG" : "Threonin", "AAG" : "Lysin",         "AGG" : "Arginin",
    "GUU" : "Valin",           "GCU" : "Alanin",   "GAU" : "Asparaginsyre", "GGU" : "Glycin",
    "GUC" : "Valin",           "GCC" : "Alanin",   "GAC" : "Asparaginsyre", "GGC" : "Glycin",
    "GUA" : "Valin",           "GCA" : "Alanin",   "GAA" : "Glutaminsyre",  "GGA" : "Glycin",
    "GUG" : "Valin",           "GCG" : "Alanin",   "GAG" : "Glutaminsyre",  "GGG" : "Glycin"
}

# Henter ut tre og tre baser fra mRNA, setter de inn i ordboken og skriver ut aminosyren.
for i in range(0, len(mRNA) - 2, 3):
  triplett = mRNA[i:i+3]
  print(kodontabell[triplett])
``````
