#### Hva er proteinsyntesen?
Proteiner er livsnødvendige stoffer som finnes i alt fra skjelett til muskler, hår, hud, blod og organer. 

Proteiner består av lange kjeder med aminosyrer. Aminosyrene fungerer litt som legoklosser, og det er i all hovedsak 20 forskjellige aminosyrer som bygger opp proteiner. Rekkefølgen og utseendet av den lange kjeden med aminosyrer er det som bestemmer egenskapene til proteinet.

Kroppen må sette sammen disse legoklossene i riktig rekkefølge for at proteinet skal gjøre det det er ment til å gjøre. 

##### DNA-molekylet 🧬
Når man bygger lego så har man ofte en slags manual, eller en oppskrift, på hvordan man skal sette sammen klossene for å få en bil, en tiger eller en Millenium Falcon™️. Hvor sitter denne *oppskriften* i kroppen? Vel, den sitter som en kode lagret i hver enkelt celle; i et molekyl som heter *DNA-molekylet* 🧬

DNA-molekylet ser ut som en dobbel spiral. Hvert "trappetrinn" består av et par baser. Det er fire baser å velge mellom i DNA: Adenin, Tymin, Cytosin og Guanin (eller bare A, T, C og G). Disse kommer alltid i par, så hvis det står A på den ene siden, så står det T på den andre. Står det C på den ene siden, så står det G på den andre siden.

Vi kan se for oss at en bit av DNA-molekylet ser slik ut:

```
TACACCTAGCATATT
ATGTGGATCGTATAA
```

DNA-molekylet befinner seg i cellekjernen. For at denne biten med DNA skal kunne bli til et protein så må den kopieres, flyttes ut og leses av.

##### mRNA og tRNA
Inne i cellekjernen kopieres en bit av DNA om til et nytt molekyl som heter mRNA. 

Selv om det er to sider til et DNA-molekyl, trenger vi bare å se på en side.

```
DNA:
TACACCTAGCATATT
```

Når mRNA skal kopiere dette segmentet så tar vi de motsatte baseparene, men basen Tymin byttes ut med Uracil.

- A -> U
- T -> A
- C -> G
- G -> C

Kopien av DNA-molekylet over ser da slik ut.

```
mRNA:
AUGUGGAUCGUAUAA
```

Dette mRNA-molekylet sendes så ut av cellekjernen og til ribosomene i cellen. En kombinasjon av tre baser kalles for en *triplett*. Hver triplett koder for en aminosyre eller et signal. Hver tRNA-bit gjenkjenner en triplett og setter riktig aminosyre der. Dermed bygges proteinet i riktig rekkefølge.

```
mRNA:
AUG UGG AUC GUA UAA

Tilsvarer aminosyrene:
Metionin(START) Tryptofan Isoleucin Valin STOPP
```

En triplett (AUG) koder for starten av proteinet. Denne koder også for metionin. Det er også tre tripletter som koder for et sluttsignal av proteinsyntesen (UAA, UAG, UGA).

Proteiner består vanligvis av hundrevis og noen ganger tusenvis av aminosyrer. Da blir det litt vanskelig å sitte å oversette base for base for å finne ut hvordan proteinet vil se ut fra en DNA-sekvens. Før vi begynner med programmering av proteinsyntesen må vi kunne litt om *ordbøker* i Python.

Ordbøker er objekter som oversetter en ting til en annen, for eksempel kan de oversette et ord til et annet.

````python
ordbok = {
    "NO" : "Norge",
    "SE" : "Sverige",
    "DK" : "Danmark"
}

print(ordbok["NO"]) # Setter inn "NO" i ordboken, og får ut "Norge"
print(ordbok["SE"]) # Setter inn "SE" i ordboken, og får ut "Sverige"
print(ordbok["DK"]) # Setter inn "DK" i ordboken, og får ut "Danmark"
````
