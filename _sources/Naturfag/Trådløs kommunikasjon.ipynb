```{admonition} Forutsetninger
:class: note
Dette opplegget fortusetter ingen tidligere programmeringskunnskaper, men kan lett gjøres enda mer utforskende og spennende om man kan litt Python.
```

For å programmerere skal vi bruke nettsiden https://python.microbit.org/v/3/.

####
