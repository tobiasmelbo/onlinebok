#### Programmere tallf√∏lger med formel

Vi kan enkelt programmere tallf√∏lger med formel. Da kan vi finne ulike summer. Vi kan ogs√• finne hvilke tall som er i f√∏lgen, og hvilke som ikke er det.

La oss finne summen av de $20$ f√∏rste kvadrattallene og unders√∏ke om tallet $289$ er et kvadrattall.

`````python
s = 0 # Summen starter som null

for n in range(1, 21): # For-l√∏kken g√•r fra 1 til √©n mindre enn den siste parameteren, alts√• 20.
    s += n**2          # Legger til summen. n**2 er formelen for kvadrattallene.
    print(n, n**2)     # Skriver ut indeks (plassering) og tall i tallf√∏lgen.

print(f"Summen er {s}")
`````

Som vi ser er tallet $289$ et kvadrattall, det er nummer 17 i f√∏lgen. I tillegg er summen av de $20$ f√∏rste kvadrattallene lik $2870$.

#### Programmere tallf√∏lger med algoritme

Det er ikke like lett √• finne en formel for alle tallf√∏lger. Noen ganger er det enklere √• programmere tallf√∏lgen fra en algoritme.

La oss se p√• *fibonaccitallene*. De er gitt ved denne algoritmen.

- Start med tallene $\{0, 1\}$
- Hvert neste tall er summen av de to forrige tallene.

La oss programmere starten p√• denne tallf√∏lgen.

`````python
fib = [0, 1]

while len(fib) < 10:
    ny_fib = fib[-1] + fib[-2] # Tar summen av det siste og det nest siste tallet i listen.
    fib.append(ny_fib)         # Legger inn tallet i listen.

print(fib)
`````

Dette programmet legger inn det neste tallet i listen som summen av de to siste tallene s√• lenge listens lengde er under 10. Derfor skriver programmet ut de f√∏rste ti fibonaccitallene.
