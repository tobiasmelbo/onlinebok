#### Programmering av vekstfaktor med `for`

La oss se p√• et eksempel. Jeg setter inn $10\,000$ kr i en bankkonto med $5\%$ √•rlig rente. Hvordan vil pengene mine se ut i de neste √•rene?

``````python
penger = 10000
prosentfaktor = 0.05
vekstfaktor = 1 + prosentfaktor

for n in range(5):
    # Skriver ut √•r og penger
    print("√Ör", n)
    print("Penger:", round(penger, 2))

    # Regner ut ny verdi for pengene ved √• gange med vekstfaktor
    penger = penger * vekstfaktor
``````

``````python
verdi = 200000
prosentfaktor = 0.09
vekstfaktor = 1 - prosentfaktor
√•r = 0

while verdi > 50000:
    # √òker √•ret
    √•r = √•r + 1

    # Minker verdien av bilen ved √• gange med vekstfaktor
    verdi = verdi * vekstfaktor

# Skriver ut verdien og √•ret
print("Etter", √•r, "√•r har bilen verdien", round(verdi, 2), "kr")
``````
