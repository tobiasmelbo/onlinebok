### L√∏kker

L√∏kker brukes for √• gjenta en del av koden flere ganger. 
 
#### `while`-l√∏kker

`while`-l√∏kker (*`while`-loops*) brukes hvis vi vet n√•r koden skal stoppe, men ikke n√∏dvendigvis hvor mange ganger den skal kj√∏re.

``````python
a = 0

while a < 5:  # Sjekker om a er mindre enn 5
    print(a)  # Skriver ut a
    a += 1    # Legger til 1 p√• a
``````

Her starter vi med `a = 1`. Deretter sjekker `while`-l√∏kken om utsagnet `a < 5` er sant. Siden det er sant s√• kj√∏res koden som st√•r med innrykk. Her skrives `a` ut og det legges til `1` p√• `a`. Dette gjentas helt til `a` er lik `5`, da blir utsagnet `a < 5` usant, og koden gjentas ikke flere ganger.

```{admonition} Viktig!
:class: warning
Hvis utsagnet (`a < 5`) aldri blir usant vil koden i `while`-l√∏kken kj√∏res "for alltid". Da krasjer programmet v√•rt!

Det er viktig √• s√∏rge for at `while`-l√∏kken har en slutt. Derfor s√∏rger vi for at `a` √∏ker i dette tilfellet.
```

(for-l√∏kker)=
#### `for`-l√∏kker

`for`-l√∏kker (*`for`-loops*) brukes hvis vi vet akkurat hvor mange ganger koden skal kj√∏res.

``````python
for n in range(5):
    print(n)
``````

Denne l√∏kken starter med `n = 0` og legger til `1` p√• `n` hver gang. Ved √• endre p√• `range()` kan vi f√• koden med innrykk til √• kj√∏re s√• mange ganger vi √∏nsker.

(for-l√∏kker med startverdi og steglengde)=
##### `for`-l√∏kker med startverdi og steglengde

Vi kan f√• mer kontroll over `for`-l√∏kker ved √• bruke flere parametre i `range()`-funksjonen.

Vi kan spesifisere en start og en slutt. Indeksene g√•r fra og med den f√∏rste verdien, men bare *til* den siste verdien. Det vil si at den siste verdien ikke blir med.

``````python
for n in range(1, 5):
    print(n) # n g√•r fra og med 1 *til* 5 (alts√• er 5 ikke med)
``````

Vi kan ogs√• spesifisere en steglengde

``````python
for n in range(1, 10, 2):
    print(n) # n g√•r fra og med 1 til 10, men med steglengde 2
``````
