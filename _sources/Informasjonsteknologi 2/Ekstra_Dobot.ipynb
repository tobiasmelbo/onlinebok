### Ekstra: Programmering av Dobot Magician 🦾
  
Dobot Magician er en robotarm som vi kan programmere i Python.

Til å bruke Dobot Magician kommer vi til å bruke programmet DobotStudio V1.9.4
 
For å kunne programmere roboten må vi vite litt om koordinatsystemet som roboten bruker.

#### Koordinatsystemet
Rydd klar en flate og orienter armen slik at strømkontakten peker utover. Jeg kommer til å referere til der strømkontakten er som *bak*.

- `x`-koordinaten bestemmer hvor langt ut armen beveger seg fra kroppen. Å øke `x` beveger armen ut fra kroppen i forhold til strømkontakten som er bak. Å minke `x` innebærer at armen beveger seg nærmere kroppen.
- `y`-koordinaten bestemmer hvor langt ut armen beveger seg langs kroppen. Å øke `y` flytter armen til venstre. Å minke `y` flytter armen til høyre.
- `z`-koordinaten bestemmer hvor langt opp eller ned armen skal være. En lav `z`-verdi putter armen nærmere bordet, mens en høy `z`-verdi løfter armen fra bordet.
- `r`-koordinaten bestemmer rotasjonen til griperverktøyet.

#### Viktige kommandoer

##### dType.SetPTPCommonParams(api, v, a, isQueued)

For å sette fart og akselerasjon bruker vi `SetPTPCommonParams`.
- `api` setter vi til `api` for alle kommandoer.
- `v` står for fart og kan være mellom `0` og `100`. 
- `a` står for akselerasjon og kan også være mellom `0` og `100`.
- `isQueued` er enten `0` eller `1` og bestemmer om kommandoen skal i en *kø*. Vi kommer til å anta at alle kommandoer vi sender skal i en kø.

```
SetPTPCommonParams(api, 50, 50, 1)
```

Det er fint å starte med halv fart og akselerasjon i begynnelsen.

##### dType.SetHOMEParams(api, x,  y, z, r, isQueued)

`HOME` er startstedet til armen. Denne kommandoen tar et sett med koordinater som vi ønsker som en startposisjon for armen.

```
SetHOMEParams(api, 145, 0, 0, 0, 1)
```

Dette er en fin `HOME`-instilling for mange bruksområder.

##### dType.GetPose(api)

Henter posisjonen til armen. Returnerer en liste `[x, y, z, r]`.

```
pos = dType.GetPose(api)
x_0 = pos[0]
y_0 = pos[1]
z_0 = pos[2]
r_0 = pos[3]
```

##### dType.SetPTPCmd(api, x, y, z, r, isQueued)

Får roboten til å bevege armen til en ny posisjon.

```
dType.SetPTPCmd(api, 2, x_0 + 10, y_0, z_0, r_0, 1) # Beveger armen 10 enheter i `x` retning.
```
