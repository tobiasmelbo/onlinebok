````{admonition} Løsning
```
def sortert(a : list):
    def bytt_plass(a : list, i : int):
        c = a[i + 1]
        a[i + 1] = a[i]
        a[i] = c

    har_byttet = True
    while har_byttet:
        har_byttet = False
        for i in range(len(a) - 1):
            if a[i] > a[i + 1]:
                bytt_plass(a, i)
                har_byttet = True
    return a
````
