### Filer
 
Vi kan både lese og skrive filer i Python.
   
#### Skrive fil ✍️

Vi kan skrive en fil på følgende måte
 
```
with open("atomer.txt", "w") as file:
    file.write("Hydrogen,H,[1],Gass \n")
    file.write("Helium,He,[2],Gass \n")
    file.write("Litium,Li,[2 1],Fast \n")
```

Kjører vi programmet over så får vi filen `atomer.txt`:

```
Hydrogen,H,[1],Gass
Helium,He,[2],Gass
Litium,Li,[2 1],Gass
```

`\n` betyr linjeskifte. Skriver vi ikke `\n` så får vi alt skrevet på samme linje.

Vi kan også skrive filer ved å bruke `for`-løkker.

```
with open("løkke.txt", "w") as file:
    for i in range(3):
        file.writeline("Linje " + str(i) + "\n")
```

Da får vi filen `løkke.txt`:

```
Linje 0
Linje 1
Linje 2
```

```{admonition} Merk deg
:class: note
Når vi skriver ting med innrykk etter `with open("løkke.txt", "w") as file:` så er filen åpen. Det er bare i dette området man har tilgang til `file`. Når man går ut fra innrykkene, så lukkes filen. 

Det finnes også andre måter å åpne og lukke filer som ikke bruker innrykk.
```

#### Legge til på en fil 📝

Tegnet `"w"` i `with open("atomer.txt", "w") as file:` står for *write*. Når vi bruker `"w"` skriver vi en fil helt fra starten av. Alt som stod der før vil bli slettet.

Hvis vi ønsker å legge til tekst på en allerede eksisterende fil kan vi bruke `"a"`. Dette står for *append*, som betyr *å legge til på slutten av*.

```
with open("atomer.txt", "a") as file:
    file.write("Sølv,Ag,[2,8,18,18,1],Fast")
```

Hvis vi kjører dette programmet ser vi at filen `atomer.txt` blir til:

```
Hydrogen,H,[1],Gass 
Helium,He,[2],Gass 
Litium,Li,[2,1],Fast 
S�lv,Ag,[2 8 18 18 1],Fast
```

Sånn kan man legge til data på en allerede eksisterende fil!

... men hvorfor blir ordet `Sølv` skrevet som `S�lv`?

#### Håndtere æ, ø og å

De nordiske bokstavene skaper noen gang problemer når vi leser og skriver filer.

Hvis vi legger til `encoding = "utf-8"` til som argument for `open()`-funksjonen så velges riktig sett med tegn.

```
with open("atomer.txt", "w", encoding = "utf-8") as file:
    file.write("Hydrogen,H,[1],Gass \n")
    file.write("Helium,He,[2],Gass \n")
    file.write("Litium,Li,[2 1],Fast \n")
    file.write("Sølv,Ag,[2 8 18 18 1],Fast")
```

Da får vi skrevet filen:

```
Hydrogen,H,[1],Gass 
Helium,He,[2],Gass 
Litium,Li,[2 1],Fast 
Sølv,Ag,[2 8 18 18 1],Fast
```

#### Lese fra fil 📖

Vi kan lese fra en fil på følgende måte

````python
with open("atomer.txt", "r", encoding = "utf-8") as file:
    linjer = file.readlines() # Alle linjene lagres i listen "linjer"

print(linjer) # Skriver ut linjene som en liste
````

Som vi ser er `linjer` en liste med strings.

Vi kan skrive ut hver enkelt linje på denne måten:

````python
with open("atomer.txt", "r", encoding = "utf-8") as file:
    linjer = file.readlines() # Alle linjene lagres i listen "linjer"

for x in linjer:
    print(x) # x er hver enkelt linje
````

Vi bruker altså `"r"` når vi ønsker å lese fra en fil. Du klarer kanskje å gjette hva `r` står for på engelsk?

##### Hoppe over linjer

Datafiler har ofte noen linjer med informasjon som ikke er selve datasettet som vi ønsker å lese.

Her er en datafil fra SSB.

```
"13931: Klimagasser AR5, etter kilde (aktivitet), komponent, år og statistikkvariabel"

"kilde (aktivitet)";"komponent";"år";"Utslipp til luft (1 000 tonn CO2-ekvivalenter, AR5)"
"1 Olje- og gassutvinning";"Klimagasser i alt";"1990";8217
"1 Olje- og gassutvinning";"Klimagasser i alt";"1991";8107
"1 Olje- og gassutvinning";"Klimagasser i alt";"1992";8710
```

Vi ønsker bare å lese av dataene fra og med linje 4. For å gjøre det kan vi hoppe over tre linjer med `file.readline()`.

````python
with open("oljegass.csv", "r", encoding = "utf-8") as file:
    file.readline() # Hopper over en linje
    file.readline() # Hopper over en linje
    file.readline() # Hopper over en linje
    linjer = file.readlines() # Lagrer resten av linjene i variablen linjer

print(linjer)
````

Som vi ser blir `linjer` en liste med strings. Hver string er en linje med data, og ingen av de første tre linjene blir med i dataene.

#### String-metoder

Nå har vi lært oss å hente ut linjer fra en datafil. Hver linje er en string. Her kommer en del metoder som er tilknyttet strings og kan være nyttige. En mer komplett oversikt kan du for eksempel finne på [W3Schools (String Methods)](https://www.w3schools.com/python/python_ref_string.asp).

Merk deg at ingen av metodene under endrer på strings, de bare returnerer nye strings som er endret på en eller annen måte. Dette er i motsetning til listemetodene `append()`, `sort()` og `reverse()` som faktisk endrer på listene når man kaller på metodene.

##### `string.split()`-metoden

Vi ønsker ofte å dele tekst opp i biter. `split()`-metoden tar en string som argument. Denne kalles en *separator*. Metoden deler teksten opp på hver separator og returnerer en liste med tekstbiter.

````python
linje = "Hydrogen,[1],Gass"

print(linje.split(",")) # Hele linjen, splittet på ","
print(linje.split(",")[0]) # Atomnavn
print(linje.split(",")[1]) # Energinivåer
print(linje.split(",")[2]) # Tilstand ved romtemperatur
````

Eksempelet over deler opp en string på `","`. 

Et annet vanlig bruksområde er å splitte på mellomrom `" "` for å få en liste med hvert ord i setninger. Vi trenger ikke en argument til `split()`-metoden for å splitte på mellomrom.

````python
setning = "Jeg elsker Python!"

print(setning.split(" ")) # Splitter på mellomrom
print(setning.split()) # Splitter også på mellomrom
````

##### `string.lower()`- og `string.upper()`-metoden

Noen ganger ønsker vi å gjøre alle bokstaver i en tekst til stor bokstav eller liten bokstav. `lower()`-metoden returnerer en tekst med små bokstaver og `upper()`-returnerer en tekst med store bokstaver.

````python
tekst = "Ja"

print(tekst.upper()) # Teksten i store bokstaver
print(tekst.lower()) # Teksten i små bokstaver
````

Dette kan være nyttig når man skal sjekke `input()` i et program.

```
inputTekst = input("Svar: ")

if inputTekst.lower() == "ja":
    print("Bra!")
```

Eksempelet over vil skrive ut `"Bra!"` uansett om man skriver `"JA"`, `"Ja"`, `"ja"` eller `"jA"`. På den måten slipper man å sjekke alle alternativene. Dette gjør et program med `input()` litt mer brukervennlig.

##### `string.replace()`-metoden

Noen ganger ønsker vi å fjerne eller bytte ut deler av en tekst. Det første argumentet til `replace()`-metoden er den teksten som skal byttes ut, det andre parametret er den teksten som det skal byttes til. `replace()`-metoden returnerer da en tekst hvor de spesifiserte bitene av teksten er byttet ut.

````python
tekst = "Jeg hater å programmere!"

print(tekst.replace("hater", "elsker")) # Bytter om alle forekomster av "hater" til "elsker"
````

Vi kan også bruke dette til å fjerne biter av tekst ved å bytte ut tekstbiten med en tom tekst.

````python
tekst = "Jeg liker ikke å programmere. Det er ikke gøy."

print(tekst.replace(" ikke", "")) # Fjerner alle biter av teksten hvor det står " ikke".
````

##### `string.strip()`-metoden

Vi kan fjerne flere tegn i starten og i slutten av en string ved å bruke `strip()`-metoden. Uten argument fjerner metoden mellomrom.

````python
tekst = "  Python er kult :)  "

print(tekst)
print(tekst.strip()) # Fjerner mellomromene på starten og slutten
````

`strip()`-metoden kan også fjerne spesifikke karakterer, enten en eller flere.

````python
tekst = " ...Python, et fantastisk språk... "

print(tekst.strip(" .")) # Fjerner alle forekomster av både mellomrom " " og ".".
````
