### Ordbøker

Ordbøker (*dictionaries*) er en måte å lagre data. I {ref}`Lister` lagrer vi data i en rekkefølge. I ordbøker lagrer vi data i par med `keys` og `values`, nøkler og verdier. En nøkkel er forbundet med sin verdi.

`````python
land = {"NO" : "Norge", "SE" : "Sverige"}

print(land["NO"])
print(land["SE"])
`````

I koden over lager vi landkoder.  `"NO"` og `"SE"` er nøklene, og `"Norge"` og `"Sverige"` er de tilhørende verdiene. Vi knytter sammen nøklene og verdiene med kolon `:`. For å hente verdier fra ordboken `land` så skriver vi inn nøkkelen med `land["NO"]` og så får vi ut verdien `"Norge"`.

Dersom vi skal sette sammen en enkel ordbok så er det vanlig å gjøre det over flere linjer. Det er litt mer oversiktlig enn eksemplet over.

`````python
land = {
    "NO" : "Norge",
    "SE" : "Sverige"
}
`````

````{admonition} Nøklene og verdiene i en ordbok trenger ikke være strings!
:class: note
Nøklene og verdiene i en ordbok kan være av mange forskjellige typer.
```
mai = {
    17 : "Nasjonaldag!",        # Nøkkelen 17 har verdien "Nasjonaldag!"
    "erVår" : True,             # Nøkkelen "erVår" har verdien True
    True : "Det er mai!"        # Nøkkelen True har verdien "Det er mai!"
}

print(mai[17])
print(mai["erVår"])
print(mai[True])
```

gir 

```
Nasjonaldag!
True
Det er mai!
```
````

#### Legge til, endre og fjerne data

For å legge til data så setter vi en ny nøkkel lik en verdi. 

`````python
land = {}                # Tom ordbok

land["NO"] = "Norge"     # Verdien "Norge" legges inn i ordboken under nøkkelen "NO"
land["DK"] = "Danmark"   # Verdien "Danmark" legges inn i ordboken under nøkkelen "DK"

print(land)
`````

For å endre på data kan vi sette en eksisterende nøkkel til en ny verdi.

`````python
land = {
    "NO" : "Norge",
    "DK" : "Danmark"
} 

print(land)

land["DK"] = "Donkey Kong" # Nøkkelen "DK" får en ny verdi "Donkey Kong" 🐒

print(land)
`````

Vi kan fjerne data ved å bruke `pop()`-metoden.

`````python
land = {
    "NO" : "Norge",
    "SE" : "Sverige",
    "DK" : "Danmark"
}

land.pop("NO") # Fjerner både nøkkel og verdi for nøkkelen "NO"

print(land)
`````

#### Løkker og ordbøker

Vi kan gå igjennom ordbøker med `for`-løkker på denne måten.

`````python
land = {
    "NO" : "Norge",
    "SE" : "Sverige"
}

for x in land:
    print("Nøkkel:", x)         # Skriver ut nøkkelen
    print("Verdi:", land[x])   # Skriver ut verdien til nøkkelen
`````

#### Lister av ordbøker

Det kan være nyttig å lage flere ordbok-objekter og lagre de i en liste.

La oss lage en liste `elevliste` som skal inneholde ordbøker med fag som nøkler og tilfeldige karakterer som verdier.

`````python
from random import randint

elevliste = []

for n in range(3):
    # Lager ordboken elev
    elev = {}

    # Legger inn karakterer
    elev["Norsk"] =  randint(1, 6)
    elev["Gym"] = randint(1, 6)
    elev["Matte"] = randint(1, 6)

    # Legger ordboken inn i listen
    elevliste.append(elev)

print(elevliste)
`````

Som vi ser har vi nå laget en liste som er fylt med ordbøker. Hvert element i listen skal representere en tilfeldig elev.
