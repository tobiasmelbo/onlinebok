### Arrays

En array er en måte å lagre data på. Arrays likner på {ref}`Lister`, men har flere viktige forskjeller.

For å bruke arrays må vi bruke en pakke. Vi bruker pakken `numpy`. Mange Python-installasjoner vil inkludere denne pakken, men man kan også installere pakken ved å skrive `pip install numpy` i terminalvinduet.

Når vi har `numpy`-pakken kan vi lage en array.

```python
import numpy as np

array = np.array([1, 2, 3])

print(array)
```

Arrays kan bare ha elementer med data av samme type.

```python
import numpy as np

array = np.array([1, "Alfa", True])
array2 = np.array([1, 2, 3.14])

print(array)
print(array2)
```

Her ser vi at alle elementene i `array` har blitt gjort om til tekststrenger, og at alle elementene i `array2` har blitt gjort om til flyttall.

En annen vesentlig forskjell er at det går mye raskere å finne elementer i arrays enn i lister, men det tar mye lengre tid å legge til og trekke ut elementer fra en array.

#### Vektorregning

Arrays er vektorer. Det gjør at vi kan bruke de til vektorregning.

```python
import numpy as np

l1 = [1, 1] # Liste 1
l2 = [1, 1] # Liste 2

a1 = np.array(l1) # Lager array av liste 1
a2 = np.array(l2) # Lager array av liste 2

print(l1 + l2) # Legger sammen lister
print(a1 + a2) # Legger sammen arrays
```
