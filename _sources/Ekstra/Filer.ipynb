### Filer

Vi kan lage og lese av filer i Python. Dette er nyttig for å lagre informasjon og for å analysere data.

#### Skrive til fil

Vi kan skrive til fil på følgende måte.

```
with open("atomer.txt", "w") as file:
    file.write("Hydrogen,[1],Gass \n")
    file.write("Helium,[2],Gass \n")
    file.write("Litium,[2,1],Fast \n")
    file.write("Beryllium,[2,2],Fast")
```

Hvis du kjører programmet over vil du få en ny fil `atomer.txt`.

```
Hydrogen,[1],Gass
Helium,[2],Gass
Litium,[2,1],Gass
Beryllium,[2,2],Fast
```

Merk at `\n` betyr linjeskifte. Skriver vi ikke `\n` så får vi alt skrevet på samme linje.

I stedet for å skrive filen linje for linje kan vi bruke en `for`-løkke.

```
with open("løkke.txt", "w"):
    for i in range(5):
        file.writeline("Linje " + i + "\n")
```

Da får vi filen `løkke.txt`.

```
Linje 0
Linje 1
Linje 2
Linje 3
Linje 4
```

#### Lese fra fil

Vi kan lese fra filer på følgende måte

```
with open("atomer.txt", "r") as file:
    linjer = file.readlines()

for x in linjer:
    print(x)
```

````python
with open("atomer.txt", "r") as file:
    linjer = file.readlines()

for x in linjer:
    print(x)
````

Noen ganger ønsker vi å hente ut deler av linjene. For å gjøre dette kan vi bruke `string.split()`-metoden. Den deler opp linjen på et tegn og lager en liste.

````python
linje = "Hydrogen,[1],Gass"

print(linje.split(",")[0]) # Atomnavn
print(linje.split(",")[1]) # Energinivåer
print(linje.split(",")[2]) # Tilstand ved romtemperatur
````
