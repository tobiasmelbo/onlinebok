### Funksjoner

Vi kan tenke på en funksjon som en oppgave som kan inneholde flere instruksjoner. For å kjøre en funksjon gjør vi et *kall* på funksjonen.  

```python
def hils():
    print("Hei!") # Skriver ut "Hei!"

hils()            # Et kall på funksjonen hils().
```

Når vi skriver `hils()` gjør vi et kall på funksjonen. Da gjennomføres det som står med innrykk.

#### Parametre og `return`-verdier

Ofte ønsker vi å sette variabler *inn* i en funksjon, la gjøre funksjonen gjøre noe, og så få *ut* noe.

Det vi setter inn kalles funksjonens **parameter**. Det vi får ut kalles funksjonens **`return`-verdi**.

```python
def dobbelt(x):   # x er navnet på funksjonens parameter
    return 2*x    # return-verdien er 2 ganger x.

print(dobbelt(2)) # 2 settes inn som parameter til funksjonen
```

Når vi kjører funksjonen med funksjonskallet `dobbelt(2)` så skjer dette:

- `2` settes inn i funksjonen som `x`
- funksjonen returnerer to ganger `x`, som blir `4`

Derfor skrives `4` ut.
