#### `while`-lû¡kker

`while`-lû¡kker (*`while`-loops*) brukes hvis vi vet nûËr koden skal stoppe, men ikke nû¡dvendigvis hvor mange ganger den skal kjû¡re.

````python
a = 0

while a < 5:  # Sjekker om a er mindre enn 5
    print(a)  # Skriver ut a
    a += 1    # Legger til 1 pûË a
````

Her starter vi med `a = 1`. Deretter sjekker `while`-lû¡kken om utsagnet `a < 5` er sant. Siden det er sant sûË kjû¡res koden som stûËr med innrykk. Her skrives `a` ut og det legges til `1` pûË `a`. Dette gjentas helt til `a` er lik `5`, da blir utsagnet `a < 5` usant, og koden gjentas ikke flere ganger.

```{admonition} Viktig!
:class: warning
Hvis utsagnet (`a < 5`) aldri blir usant vil koden i `while`-lû¡kken kjû¡res "for alltid". Da krasjer programmet vûËrt!

Det er viktig ûË sû¡rge for at `while`-lû¡kken har en slutt. Derfor sû¡rger vi for at `a` û¡ker i dette tilfellet.
```

#### `for`-lû¡kker

`for`-lû¡kker (*`for`-loops*) brukes hvis vi vet akkurat hvor mange ganger koden skal kjû¡res.

````python
for n in range(5):
    print(n)
````

Denne lû¡kken starter med `n = 0` og legger til `1` pûË `n` hver gang. Ved ûË endre pûË `range()` kan vi fûË koden til ûË kjû¡re sûË mange ganger vi û¡nsker.
