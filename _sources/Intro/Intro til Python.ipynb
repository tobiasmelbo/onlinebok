# Intro til Python

I denne delen vil jeg introdusere deg for programmering i Python.

Du vil lære flere grunnleggende ting slik at du kan

- lage programmer for å modellere i naturfag 🦠
- simulere i fysikk 🚀 
- bruke numeriske metoder for å løse matematiske problemer 🧮
- ... eller bare å lære deg å programmere 🧑‍💻

Det er mange måter å lage og kjøre programmer i Python. Jeg anbefaler å bruke [Trinket](https://trinket.io/python3) for å programmere enkelt i nettleseren eller et program som [VSCode](https://code.visualstudio.com/docs/python/python-tutorial) om du skal programmere mye.
