### Input og pakker

#### Pakker

##### `math`-pakken

Pakken `math` gir oss flere nyttige verktøy. 

- `math.sqrt(<tall>)` er en funksjon som gir oss kvadratroten av et tall.
- `math.pi` gir oss $\pi$.

Her er noen eksempler.

````python
import math

print(math.sqrt(4))     # Gir oss kvadratroten av 4.
print(math.pi)          # Gir oss pi.
````

Man kan også importere bare de funksjonene man trenger. Det kan gjøre koden litt enklere.

````python
from math import sqrt, pi

print(sqrt(4))
print(pi)
````

(random-pakken)=
##### `random`-pakken

Pakken `random` gir oss tilfeldige tall. Ikke bare er det gøy å leke med, men det er også nyttig for å simulere ting.

- `random.random()` er en funksjon som gir oss et tilfeldig tall fra og med null og opp til 1. Altså $[0,1\rangle$.
- `random.randint(<start>, <slutt>)` er en funksjon som gir oss et tifeldig tall fra og med startverdien til og med sluttverdien.

Her er noen eksempler

````python
from random import random, randint

print(random())      # Gir oss et tilfeldig tall fra og med null og opp til 1.
print(randint(1,6))  # Gir oss et tall fra og med 1 til og med 6 (akkurat som et terningkast).
````

#### Input med `input()`-funksjonen

Vi ønsker noen ganger å ta input fra brukeren. Det kan vi gjøre med `input()`-funksjonen.

````python
navn = input("Navn:")
print("Hei,", navn)
````

#### Avrunding med `round()`-funksjonen

Vi kan runde av lange desimaltall ved å bruke `round`-funksjonen. Denne ligger ikke i en pakke og må derfor ikke importeres.

````python
from math import pi

print(pi)           # Skriver ut pi med mange desimaler
print(round(pi, 2)) # Skriver ut pi rundet av til to desimaler.
````
