### Variabler og syntaks

En variabel inneholder verdier. Vi lager en variabel ved å gi den et navn og en verdi. Dette gjør vi med likhetstegnet `=` i Python.  

```
a = 2
b = 3
c = a + b
```

#### Variabeltyper

Det finnes mange typer variabler. Vi skal gå i detalj på flere av disse senere.

```
a = 1                   # Heltall (integer)
b = 2.3                 # Flyttall eller desimaltall (float)
c = "Hei!"              # Streng (string)
d = True                # Sannhetsverdi (boolean)
e = [2, 3]              # Liste (list)
f = {"NO" : "Norway"}   # Ordbok (dictionary)
```

```{admonition} Merk deg
:class: note
I Python bruker vi punktum `.` for desimaltall, ikke komma `,`.
```

#### Kommentarer

Noen ganger ønsker vi å legge inn kommentarer i koden for å gjøre den mer lesbar. 

```
# Dette er en kommentar på en linje.

"""
Dette er en kommentar over flere linjer.
Vi kan skrive så mye vi vil her inne.
"""
```

#### `print()`-funksjonen

Vi ønsker ofte at programmet skal skrive ut noe. Til det bruker vi ```print()```-funksjonen.

````python
a = 2
b = 3
c = a + b

print(c)
````

`print()`-funksjonen kan også skrive ut tekst eller flere variabler med mellomrom dersom vi skiller variablene med komma.

````python
a = 2
b = 3
c = a + b

print(a, "pluss", b, "er lik", c)
````
