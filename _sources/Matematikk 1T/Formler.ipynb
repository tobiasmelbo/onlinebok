### Formler

I matematikken er formler sammenhenger som blir uttrykt med matematisk språk. Vi bruker ofte formler til å regne på ulike størrelser i virkeligheten og i geometrien. Formler er også viktige når vi snakker om figurtall.

```{admonition} Forutsetninger
:class: note
Denne siden forutsetter at man har vært gjennom delkapitlene

- {ref}`Kalkulatorer og pakker`
- {ref}`Vilkår`
- {ref}`Løkker`
```

#### Eksempler på formler.
Her kommer noen eksempler på formler fra geometrien, fysikk og naturfag.

##### Plangeometri
::::{grid} 3
:gutter: 1

:::{grid-item-card} Navn
:::

:::{grid-item-card} Areal
:::

:::{grid-item-card} Omkrets

:::

:::{grid-item-card}
Sirkel
:::

:::{grid-item-card}
$\pi r^2$
:::

:::{grid-item-card}
$2\pi r$
:::

:::{grid-item-card}
Trekant
:::

:::{grid-item-card}
$\frac{1}{2}gh$
:::

:::{grid-item-card}
$a+b+c$
:::

:::{grid-item-card}
Rektangel
:::

:::{grid-item-card}
$a\cdot b$
:::

:::{grid-item-card}
$2a + 2b$
:::

::::

##### Romgeometri 
::::{grid} 3
:gutter: 1

:::{grid-item-card} Navn
:::

:::{grid-item-card} Volum
:::

:::{grid-item-card} Overflate

:::

:::{grid-item-card}
Kule
:::

:::{grid-item-card}
$\frac{4}{3}\pi r^3$
:::

:::{grid-item-card}
$4\pi r^2$
:::

:::{grid-item-card}
Sylinder
:::

:::{grid-item-card}
$\pi r^2h$
:::

:::{grid-item-card}
$2\pi r^2 + 2\pi rh$
:::

:::{grid-item-card}
Kjegle
:::

:::{grid-item-card}
$\frac{1}{3}\pi r^2h$
:::

:::{grid-item-card}
$\pi r^2 + \pi rs$
:::

::::

#### Formler for figurtall.
