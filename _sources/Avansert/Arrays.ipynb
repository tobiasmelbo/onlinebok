### Arrays

En array er en måte å lagre data på. Arrays likner på {ref}`Lister`, men har flere viktige forskjeller.

For å bruke arrays må vi bruke en pakke som heter `numpy`. Mange Python-installasjoner vil inkludere denne pakken, men man kan også installere pakken ved å skrive `pip install numpy` i terminalvinduet.

Når vi har `numpy`-pakken kan vi lage en array.

```python
import numpy as np

array = np.array([1, 2, 3])

print(array)
```

Arrays kan bare ha elementer med data av samme type. Hvis man legger inn flere typer data vil alle blir gjort om til samme datatype.

```python
import numpy as np

array = np.array([1, "Alfa", True])
array2 = np.array([1, 2, 3.14])

print(array)
print(array2)
```

Her ser vi at alle elementene i `array` har blitt gjort om til tekststrenger, og at alle elementene i `array2` har blitt gjort om til flyttall. Dette kalles *upcasting*. Vi trenger ikke tenke på dette om vi bare jobber med samme type elementer i et array.

#### Vektorregning

Arrays er vektorer. Det gjør at vi kan bruke de til vektorregning.

```python
import numpy as np

l1 = [1, 1] # Liste 1
l2 = [1, 1] # Liste 2

a1 = np.array(l1) # Lager array av liste 1
a2 = np.array(l2) # Lager array av liste 2

print(l1 + l2) # Legger sammen lister
print(a1 + a2) # Legger sammen arrays
```

Legger vi sammen lister, så settes de sammen til en ny større liste. Altså blir `[1, 1] + [1, 1] = [1, 1, 1, 1]`. 

Legger vi sammen arrays så fungerer det på samme måte som i vektorregningen. Altså at $[1, 1] + [1, 1] = [1+1, 1+1] = [2, 2]$

#### `linspace()`-funksjonen

For å lage en jevn fordeling av tall i et intervall kan vi bruke `linspace()`-funksjonen. Denne er veldig nyttig for å lage `x`-verdier til en funksjon.

```python
import numpy as np

print(np.linspace(0, 2, 5)) # 5 tall mellom 0 og 2 (inkluderende)
print(np.linspace(-5, 5, 10)) # 10 tall mellom -5 og 5 (inkluderende)
```

Her kommer et eksempel på hvordan vi kan lage verdier med `linspace()` og bruke det til å finne tilsvarende `y`-verdier.

```python
import numpy as np

def f(x):
    return x**2

x = np.linspace(-2, 2, 5) # 50 tall mellom -5 og 5
y = f(x)

print("x-verdier:", x)
print("y-verdier:", y)
```
